```python
import pandas as pd
import requests
from pprint import pprint
import json
from datetime import datetime
```

```python
url = f"https://api.fda.gov/drug/enforcement.json?limit=500"
```

```python
response = requests.get(url)
data = response.json()
pprint(data)
```

```python

recall_data = pd.DataFrame(data["results"])
pd.set_option('display.max_columns', None)
recall_data.head()
```

```python
recall_data.info()
```

```python
route_list = []
substance_name_list = []
for i in range(len(data["results"])):
    myitem = data["results"][i]["openfda"]
    if "route" in myitem:
        route_list.append(myitem["route"])
    else:
        route_list.append(["na"])
    if "substance_name" in myitem:
        substance_name_list.append(myitem["substance_name"])
    else:
        substance_name_list.append(["na"])
print(substance_name_list)
```

```python
route = []
for i in route_list:
    route.append(f'{i[0]}')
substance_name = []
for i in substance_name_list:
    substance_name.append(f'{i[0]}')
```

```python
recall_data = recall_data.drop(['address_1','address_2','code_info','openfda','recall_number','report_date','more_code_info','product_type','center_classification_date'],axis =1)
```

```python
recall_data["Ingestion Route"] = route
recall_data["Substance Name"] = substance_name_list
```

```python
recall_data.head()
```

```python
recall_data = recall_data.rename(columns = {"event_id":"Event ID", "recalling_firm":"Recalling Firm", "product_description":"Product Description", 
                                            "reason_for_recall":"Reason", "classification":"Classification", "product_quantity":"Quantity",
                                            "voluntary_mandated":"Voluntary/Mandated", "distribution_pattern":"Distribution", 
                                            "initial_firm_notification":"Initial Notification Method", "recall_initiation_date":"Recall Begins",
                                            "termination_date":"Recall End", "status":"Status", "city":"City", "state":"State", 
                                            "postal_code": "Postal Code", "country":"Country"})
```
